```python
import pandas as pd

df = pd.read_csv("/content/gender_submission (1).csv")
display(df.head())
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())

print("\nDataFrame Information:")
df.info()

print("\nDescriptive Statistics for numerical columns:")
print(df.describe())
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

survival_counts = df['Survived'].value_counts()
survival_proportions = df['Survived'].value_counts(normalize=True)

print("Survival Counts:")
print(survival_counts)
print("\nSurvival Proportions:")
print(survival_proportions)

# Plotting the survival distribution
plt.figure(figsize=(6, 4))
survival_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.xticks(ticks=[0, 1], labels=['Not Survived (0)', 'Survived (1)'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
